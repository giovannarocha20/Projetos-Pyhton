## Atividade 1
Primeiro, vamos criar o DataFrame com as informações.

```python
import pandas as pd

data = {
    "Sede": ["Nova Iorque", "São Paulo", "Nova Iorque", "Londres", "Londres"],
    "Piloto": ["Mike Ross", "Sebastian Thomas", "Glen Are", "Michael Schum", "Luiz da Silva"],
    "Mundiais": [10, 11, 0, 3, 44],
    "Vitórias": [321, 229, 12, 44, 1023]
}

index_labels = ["X Racing", "Equatorial", "Typo", "Blue Race", "Capa Racing"]

df = pd.DataFrame(data, index=index_labels)

df
```

## Atividade 2
Mostre todas as informações da equipe "Blue Race".

```python
df.loc["Blue Race"]
```

## Atividade 3
Mostre o número de mundiais da equipe "Capa Racing".

```python
df.loc["Capa Racing", "Mundiais"]
```

## Atividade 4
Mostre apenas as equipes que têm 10 ou mais mundiais.

```python
df[df["Mundiais"] >= 10]
```

## Atividade 5
Mostre apenas as equipes que têm 10 ou mais mundiais e mais de 300 vitórias.

```python
df[(df["Mundiais"] >= 10) & (df["Vitórias"] > 300)]
```

## Atividade 6
Atualize o nome do piloto da equipe "Equatorial" para "Antônio Racer".

```python
df.loc["Equatorial", "Piloto"] = "Antônio Racer"
df
```

## Atividade 7
Atualize a equipe "X Racing", definindo "Sede" como "São Paulo" e "Vitórias" como 322.

```python
df.loc["X Racing", ["Sede", "Vitórias"]] = ["São Paulo", 322]
df
```

## Atividade 8
Inclua uma nova equipe "Red Cow", com sede em "São Paulo", piloto "Fernando Vetel", mundiais zero e vitórias zero.

```python
df.loc["Red Cow"] = ["São Paulo", "Fernando Vetel", 0, 0]
df
```

## Atividade 9
Ordene o DataFrame de forma decrescente pelo número de vitórias.

```python
df_sorted = df.sort_values(by="Vitórias", ascending=False)
df_sorted
```
